```python
import matplotlib.pyplot as plt
import numpy as np

# Dados exatos extraídos dos SEUS mapas (Região de Igarapé-Miri/Abaetetuba/Cametá)
estacoes = ['Verão\n(Dez-Fev)', 'Outono\n(Mar-Mai)', 'Inverno\n(Jun-Ago)', 'Primavera\n(Set-Nov)']

precipitacao = [250, 420, 100, 50]         # mm
temperatura = [28.5, 26.5, 29.0, 32.0]      # °C
umidade_especifica = [0.016, 0.018, 0.0145, 0.0135] # kg/kg
vento = [3.0, 2.0, 4.0, 5.0]               # m/s

# Índices conceituais para análise integrada (Escala de 1 a 5)
indice_producao = [1, 2, 4, 5]
atividade_barbeiro = [3, 2, 4, 5]

print("Dados carregados com sucesso! Siga para a próxima célula.")
```

```python
# Criando a estrutura com 2 subplots verticais (Gráfico de Clima em cima, Biologia embaixo)
fig, (ax1, ax3) = plt.subplots(2, 1, figsize=(11, 10), dpi=300, sharex=True)

# --- SUBPLOT 1: DINÂMICA CLIMÁTICA ---
ax2 = ax1.twinx()
# Plotando as linhas com marcadores para cada variável climática
p1, = ax1.plot(estacoes, temperatura, color='darkred', marker='o', linewidth=2.5, label='Temperatura (°C)')
p2, = ax1.plot(estacoes, precipitacao, color='blue', marker='^', linewidth=2, linestyle=':', label='Precipitação (mm)')
p3, = ax2.plot(estacoes, umidade_especifica, color='darkgreen', marker='s', linewidth=2, label='Umidade Esp. (kg/kg)')

ax1.set_ylabel('Temperatura (°C) / Precipitação (mm)', color='black', fontsize=11)
ax2.set_ylabel('Umidade Específica (kg/kg)', color='darkgreen', fontsize=11)
ax1.set_title('Acompanhamento Climático Sazonal (ERA5 - 2024 às 15h)', fontsize=12, fontweight='bold')
ax1.grid(True, linestyle='--', alpha=0.5)

# Organizando as legendas do primeiro gráfico
linhas_id1 = [p1, p2, p3]
ax1.legend(linhas_id1, [l.get_label() for l in linhas_id1], loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=3)


# --- SUBPLOT 2: ANÁLISE INTEGRADA (AÇAÍ E VETOR) ---
# Barras para representar o volume da Safra de Açaí
barras = ax3.bar(estacoes, indice_producao, color='#7A4F97', alpha=0.6, width=0.35, label='Índice de Produção de Açaí')
# Linha para a atividade teórica do Barbeiro
p4, = ax3.plot(estacoes, atividade_barbeiro, color='red', marker='X', linewidth=2.5, label='Atividade Estimada do Barbeiro')

# Eixo secundário para a velocidade do vento
ax4 = ax3.twinx()
p5, = ax4.plot(estacoes, vento, color='teal', marker='p', linewidth=2, linestyle='--', label='Intensidade do Vento (m/s)')

ax3.set_ylabel('Índice de Produção / Atividade (1 a 5)', color='#7A4F97', fontsize=11)
ax4.set_ylabel('Velocidade do Vento (m/s)', color='teal', fontsize=11)
ax3.set_ylim(0, 6)
ax3.set_title('Correlação Ecológica: Safra do Açaí vs. Atividade do Barbeiro', fontsize=12, fontweight='bold')
ax3.grid(True, linestyle='--', alpha=0.5)

# Organizando as legendas do segundo gráfico (AQUI ESTAVA O ERRO CORRIGIDO: linhas2)
linhas_id2 = [barras, p4, p5]
ax3.legend(linhas_id2, [l.get_label() for l in linhas_id2], loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=3)

# Título principal do trabalho
plt.suptitle('Parte 3 - Análise Integrada Sazonal da Região Produtora de Açaí (PA)', fontsize=15, fontweight='bold', y=0.96)
plt.tight_layout(rect=[0, 0, 1, 0.93])

# EXIBE O GRÁFICO NA TELA DO COLAB
plt.show()
```
