### Visualizando a Evapotranspiração no Sudeste do Brasil

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import numpy as np

# Convert the MultiIndex DataFrame to a regular DataFrame for easier filtering
df_plot = df_converted.reset_index()

# Define the bounding box for the Southeast region of Brazil
# (Min_lat, Max_lat, Min_lon, Max_lon)
# Longitudes are in 0-360, so convert W longitudes to this format
southeast_bounds = {
    'min_lat': -26.0,  # Southern limit of São Paulo
    'max_lat': -17.0,  # Northern limit of Minas Gerais
    'min_lon': 360 - 52.0,  # Western limit of São Paulo (-52.0 W -> 308.0)
    'max_lon': 360 - 39.0   # Eastern limit of Espírito Santo (-39.0 W -> 321.0)
}

# Filter the DataFrame to the Southeast region
df_southeast = df_plot[
    (df_plot['latitude'] >= southeast_bounds['min_lat']) &
    (df_plot['latitude'] <= southeast_bounds['max_lat']) &
    (df_plot['longitude'] >= southeast_bounds['min_lon']) &
    (df_plot['longitude'] <= southeast_bounds['max_lon']) &
    (df_plot['pev'].notna()) # Filter out NaN values for plotting
]

# If multiple valid_time entries exist, take the mean for simplicity in plotting.
# For this dataset, it appears to be a single time point, but this makes it robust.
df_southeast_avg = df_southeast.groupby(['latitude', 'longitude'])['pev'].mean().reset_index()

# Prepare data for plotting
plot_lons = df_southeast_avg['longitude'].values
plot_lats = df_southeast_avg['latitude'].values
plot_pev = df_southeast_avg['pev'].values

# Create figure
fig = plt.figure(figsize=(12, 10))

# Map projection
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())

# Add map features
ax.add_feature(cfeature.LAND, edgecolor='black', zorder=0)
ax.add_feature(cfeature.OCEAN, zorder=0)
ax.add_feature(cfeature.COASTLINE, zorder=1)
ax.add_feature(cfeature.BORDERS, linestyle=':', edgecolor='gray', zorder=1)
ax.add_feature(cfeature.STATES, linestyle=':', edgecolor='darkgray', zorder=1)

# Set extent to the Southeast region
# Need to convert back to standard -180 to 180 for cartopy.set_extent
extent_min_lon = southeast_bounds['min_lon'] - 360 if southeast_bounds['min_lon'] > 180 else southeast_bounds['min_lon']
extent_max_lon = southeast_bounds['max_lon'] - 360 if southeast_bounds['max_lon'] > 180 else southeast_bounds['max_lon']

ax.set_extent([extent_min_lon, extent_max_lon,
               southeast_bounds['min_lat'], southeast_bounds['max_lat']],
              crs=ccrs.PlateCarree())

# Plot evapotranspiration data
# Longitudes for scatter also need to be converted for plotting on a -180 to 180 map
plot_lons_converted = np.where(plot_lons > 180, plot_lons - 360, plot_lons)

scatter = ax.scatter(
    plot_lons_converted,
    plot_lats,
    c=plot_pev,
    cmap='viridis',
    s=10, # Adjust point size as needed
    transform=ccrs.PlateCarree(),
    zorder=2
)

# Add colorbar
cbar = plt.colorbar(scatter, ax=ax, orientation='vertical', pad=0.05, shrink=0.7)
cbar.set_label('Potential Evaporation (pev)')

# Add gridlines and labels
gl = ax.gridlines(draw_labels=True, linestyle='--', color='gray', alpha=0.6)
gl.top_labels = False
gl.right_labels = False

# Title
plt.title('Potential Evaporation in Southeast Brazil (2024-02-01)')

plt.show()
```

### Gráfico Comparativo de Média Mensal de Evapotranspiração e Anomalia de Precipitação para Igarapé-Miri (Fevereiro de 2024)

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.patches as mpatches
import numpy as np

if 'df_precip_plot' in locals() and not df_precip_plot.empty:
    # As coordenadas para Igarapé-Miri (conforme definido anteriormente)
    coordenadas = {
        'lat_sul': -2.25,
        'lat_norte': -1.75,
        'lon_oeste': -49.20,
        'lon_leste': -48.70
    }

    # Convertendo as longitudes para o formato 0-360 para filtrar o DataFrame
    # if the data is in 0-360 longitude format
    igarapemiri_bounds_360 = {
        'min_lat': coordenadas['lat_sul'],
        'max_lat': coordenadas['lat_norte'],
        'min_lon': 360 + coordenadas['lon_oeste'],
        'max_lon': 360 + coordenadas['lon_leste']
    }

    # Check if precipitation longitude data is in 0-360 format or -180 to 180
    # Assuming 'longitude' column contains values that need conversion if < 0
    # Adjust igarapemiri_bounds_360 min_lon and max_lon if data is not 0-360
    if df_precip_plot['longitude'].min() < 0:
        igarapemiri_min_lon_filter = coordenadas['lon_oeste']
        igarapemiri_max_lon_filter = coordenadas['lon_leste']
    else:
        igarapemiri_min_lon_filter = igarapemiri_bounds_360['min_lon']
        igarapemiri_max_lon_filter = igarapemiri_bounds_360['max_lon']


    # Filtrar o DataFrame para a região de Igarapé-Miri
    df_igarapemiri_precip = df_precip_plot[
        (df_precip_plot['latitude'] >= igarapemiri_bounds_360['min_lat']) &
        (df_precip_plot['latitude'] <= igarapemiri_bounds_360['max_lat']) &
        (df_precip_plot['longitude'] >= igarapemiri_min_lon_filter) &
        (df_precip_plot['longitude'] <= igarapemiri_max_lon_filter) &
        (df_precip_plot['tp_anomaly'].notna()) # Filtrar valores NaN para plotagem (coluna corrigida para 'tp_anomaly')
    ]

    # Agrupar e obter a média se houver múltiplos pontos de tempo para a mesma coordenada
    df_igarapemiri_precip_avg = df_igarapemiri_precip.groupby(['latitude', 'longitude'])['tp_anomaly'].mean().reset_index()

    # Preparar dados para plotagem
    plot_lons_precip = df_igarapemiri_precip_avg['longitude'].values
    plot_lats_precip = df_igarapemiri_precip_avg['latitude'].values
    plot_anomaly = df_igarapemiri_precip_avg['tp_anomaly'].values

    # Criar figura
    fig = plt.figure(figsize=(12, 10))

    # Projeção do mapa
    ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())

    # Adicionar elementos geográficos detalhados (com '10m' de escala para alta resolução)
    ax.add_feature(cfeature.LAND.with_scale('10m'), facecolor='#f5f5dc', edgecolor='black', zorder=1) # Mais claro
    ax.add_feature(cfeature.OCEAN.with_scale('10m'), facecolor='#a0c8f0', zorder=1) # Azul claro
    ax.add_feature(cfeature.COASTLINE.with_scale('10m'), linewidth=1.5, zorder=2)
    ax.add_feature(cfeature.BORDERS.with_scale('10m'), linestyle=':', edgecolor='gray', zorder=2)
    ax.add_feature(cfeature.STATES.with_scale('10m'), linestyle='-', edgecolor='darkgray', zorder=2) # Linha sólida
    ax.add_feature(cfeature.RIVERS.with_scale('10m'), zorder=2, edgecolor='blue', linewidth=0.8)

    # Ajustar o ZOOM para a área de Igarapé-Miri (convertendo de volta para -180 a 180)
    # The coordinates for `set_extent` are always in -180 to 180 range
    ax.set_extent([
        coordenadas['lon_oeste'] - 0.2,
        coordenadas['lon_leste'] + 0.2,
        coordenadas['lat_sul'] - 0.2,
        coordenadas['lat_norte'] + 0.2
    ], crs=ccrs.PlateCarree())

    # Desenhar o retângulo da área selecionada (somente contorno)
    largura_lon = coordenadas['lon_leste'] - coordenadas['lon_oeste']
    altura_lat = coordenadas['lat_norte'] - coordenadas['lat_sul']

    retangulo = mpatches.Rectangle(
        (coordenadas['lon_oeste'], coordenadas['lat_sul']),
        largura_lon, altura_lat,
        linewidth=2.5, edgecolor='red', facecolor='none', # Mais espesso
        transform=ccrs.PlateCarree(), zorder=4
    )
    ax.add_patch(retangulo)

    # Adicionar um ponto central para destacar ainda mais a área
    lon_centro = (coordenadas['lon_oeste'] + coordenadas['lon_leste']) / 2
    lat_centro = (coordenadas['lat_sul'] + coordenadas['lat_norte']) / 2
    ax.plot(lon_centro, lat_centro, 'o', color='gold', markersize=10, transform=ccrs.PlateCarree(), zorder=5, markeredgecolor='black') # Ponto dourado com borda preta

    # Plotar os dados de anomalia de precipitação
    # As longitudes para o scatter também precisam ser convertidas para o intervalo -180 a 180
    plot_lons_converted_precip = np.where(plot_lons_precip > 180, plot_lons_precip - 360, plot_lons_precip)

    scatter = ax.scatter(
        plot_lons_converted_precip,
        plot_lats_precip,
        c=plot_anomaly,
        cmap='RdBu_r', # Um cmap comum para anomalias (vermelho para positivo, azul para negativo)
        s=70, # Aumentar o tamanho dos pontos
        transform=ccrs.PlateCarree(),
        zorder=3
    )

    # Adicionar barra de cores
    cbar = plt.colorbar(scatter, ax=ax, orientation='vertical', pad=0.05, shrink=0.7)
    cbar.set_label('Anomalia de Precipitação (mm/mês)', fontsize=12)
    cbar.ax.tick_params(labelsize=10)

    # Add custom text legends for colorbar interpretation
    cbar_pos = cbar.ax.get_position()
    fig.text(cbar_pos.x0 + cbar_pos.width / 2, cbar_pos.y1 + 0.02,
             'Mais precipitação\nque a média', ha='center', va='bottom', fontsize=10, color='red')
    fig.text(cbar_pos.x0 + cbar_pos.width / 2, cbar_pos.y0 - 0.02,
             'Menos precipitação\nque a média', ha='center', va='top', fontsize=10, color='blue')

    # Adicionar linhas de grade e rótulos
    gl = ax.gridlines(draw_labels=True, linestyle='--', color='gray', alpha=0.7, linewidth=0.7) # Mais visíveis
    gl.top_labels = False
    gl.right_labels = False
    gl.xlabels_top = False # Remover rótulos superiores
    gl.ylabels_right = False # Remover rótulos da direita

    # Adicionar um indicador de Norte
    ax.text(0.95, 0.95, 'N', transform=ax.transAxes,
            fontsize=16, color='black', ha='right', va='top',
            bbox=dict(boxstyle='round,pad=0.2', fc='white', alpha=0.7, ec='none'), zorder=6)


    # Título
    plt.title('Anomalia de Precipitação em Igarapé-Miri (Fevereiro de 2024)', fontsize=16, pad=15)
    plt.tight_layout()
    plt.show()
else:
    print('DataFrame de precipitação vazio ou não encontrado. Não foi possível gerar o mapa.')
```

```python
# ==============================================================================
# 1. INSTALAÇÃO DAS BIBLIOTECAS NECESSÁRIAS
# ==============================================================================
!pip install -q cdsapi xarray netCDF4 cartopy matplotlib

import cdsapi
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import numpy as np

# ==============================================================================
# 2. CONFIGURAÇÃO DA REQUISIÇÃO (CDS API)
#    -> ATENÇÃO: substitua sua chave abaixo apenas localmente. Depois de usar,
#       regenere a chave no seu perfil do CDS por segurança, já que foi
#       exposta em texto puro nesta conversa.
# ==============================================================================
SUA_CHAVE_CDS = 'a4e8e82f-5af7-4db4-8b86-00ed6386a40a'   # <<< SUBSTITUA AQUI PELA SUA CHAVE REAL DO CDS API

client = cdsapi.Client(
    url='https://cds.climate.copernicus.eu/api',
    key=SUA_CHAVE_CDS,
    verify=False
)

dataset = "reanalysis-era5-land-monthly-means"
request = {
    "product_type": ["monthly_averaged_reanalysis"],
    "variable": ["2m_temperature"],
    "year": ["2023"],
    "month": ["03"],
    "time": ["00:00"],
    "data_format": "netcdf",
    "download_format": "unarchived",
    # Área: [Norte, Oeste, Sul, Leste] -> mesma região de estudo do açaí
    "area": [-1.75, -49.2, -2.25, -48.7]
}

print("⬇️ Baixando dados do ERA5-Land (temperatura, março/2023)...")
resultado = client.retrieve(dataset, request)
caminho_arquivo = resultado.download()
print(f"✅ Arquivo baixado: {caminho_arquivo}")

# ==============================================================================
# 3. ABRINDO E PROCESSANDO O NETCDF
# ==============================================================================
ds = xr.open_dataset(caminho_arquivo)
print("\n🔎 Variáveis disponíveis no arquivo:")
print(list(ds.data_vars))

# A variável de temperatura em geral vem como 't2m' (Kelvin)
temp_kelvin = ds['t2m']
temp_celsius = temp_kelvin - 273.15   # conversão K -> °C

# Remove dimensão de tempo (só temos 1 mês), pega a matriz 2D lat/lon
temp_celsius_2d = temp_celsius.isel(valid_time=0) if 'valid_time' in temp_celsius.dims else temp_celsius.squeeze()

# ==============================================================================
# 4. COORDENADAS DA ÁREA DE ESTUDO (retângulo vermelho)
# ==============================================================================
lat_north, lat_south = -1.75, -2.25
lon_west, lon_east = -49.20, -48.70

# ==============================================================================
# 5. PLOTAGEM DO MAPA DE TEMPERATURA
# ==============================================================================
fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection=ccrs.PlateCarree())

# Define a extensão do mapa um pouco maior que a área de estudo
ax.set_extent([lon_west - 0.3, lon_east + 0.3, lat_south - 0.3, lat_north + 0.3],
               crs=ccrs.PlateCarree())

# Camada de temperatura (heatmap)
mapa_temp = temp_celsius_2d.plot(
    ax=ax,
    transform=ccrs.PlateCarree(),
    cmap='RdYlBu_r',        # azul (frio) -> vermelho (quente)
    cbar_kwargs={'label': 'Temperatura média (°C) - Março 2023'},
    add_colorbar=True
)

# Contorno da área de estudo (retângulo vermelho)
from matplotlib.patches import Rectangle
retangulo = Rectangle(
    (lon_west, lat_south),
    lon_east - lon_west,
    lat_north - lat_south,
    linewidth=2.5,
    edgecolor='red',
    facecolor='none',
    transform=ccrs.PlateCarree(),
    zorder=5
)
ax.add_patch(retangulo)

# Elementos de referência geográfica
ax.add_feature(cfeature.COASTLINE, linewidth=0.8)
ax.add_feature(cfeature.RIVERS, linewidth=0.6, edgecolor='blue')
ax.add_feature(cfeature.BORDERS, linestyle=':', linewidth=0.5)
gl = ax.gridlines(draw_labels=True, linewidth=0.3, alpha=0.5)
gl.top_labels = False
gl.right_labels = False

plt.title('Temperatura Média a 2m - ERA5-Land\nMarço/2023 - Região do Açaí (PA)', fontsize=13)
plt.tight_layout()
plt.savefig('mapa_temperatura_acai_2023.png', dpi=200, bbox_inches='tight')
plt.show()

print("\n📊 Estatísticas da área:")
print(f"  Temperatura média: {float(temp_celsius_2d.mean()):.2f} °C")
print(f"  Temperatura mínima: {float(temp_celsius_2d.min()):.2f} °C")
print(f"  Temperatura máxima: {float(temp_celsius_2d.max()):.2f} °C")
```
